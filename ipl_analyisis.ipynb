```python
class toss_win_ratio:
    def __init__(self,team_name,year):
        self.team_name=team_name
        self.year=year
    def overall_plot(self):
        print(f"All over Toss to win ratio of ",self.team_name)
        # Ploting for all seasons
        toss_won_rcb =len(data_match_08to20 [data_match_08to20["toss_winner"]==self.team_name])
        match_won_after_toss=len(data_match_08to20 [(data_match_08to20["toss_winner"]==self.team_name) &( data_match_08to20["winner"]==self.team_name)])
        toss_match_won_ratio=[match_won_after_toss,toss_won_rcb-match_won_after_toss]
        label_rcb=["Match Won","Match Lost"]
        palette_color = sns.color_palette('Set2')
        plt.pie(toss_match_won_ratio,labels=label_rcb,colors=palette_color, autopct='%.0f%%',explode=(0.1,0))
        plt.title(f"Toatl Toss  won  by "+self.team_name)
        plt.show();
        label=[self.team_name]
        x = np.arange(len(label)) # the label locations
        width = 0.35  # the width of the bars
        fig, ax = plt.subplots()
        rects1 = ax.bar(x - width/2, toss_won_rcb, width, label='Toss won')
        rects2 = ax.bar(x + width/2, match_won_after_toss, width, label='Matches won')
        # Add some text for labels, title and custom x-axis tick labels, etc.
        ax.set_ylabel('Number')
        ax.set_title(f"Toss to matches won ratio of "+self.team_name)
        ax.set_xticks(x, label)
        ax.legend()
        ax.bar_label(rects1, padding=3)
        ax.bar_label(rects2, padding=3)
        fig.tight_layout()
        plt.show()
        #Ploting for coustum years
    def season_by_season_plot(self):

        year=self.year
        toss_won=[]
        Year=[]  
        
        for i in year:
            

            result =data_match_08to20 [(data_match_08to20["toss_winner"]==self.team_name)&(data_match_08to20["year"]==i)]
            print(f"\n For Season ",i,"\n")
            if(i in result.values):
                toss_won_rcb =len(data_match_08to20 [(data_match_08to20["toss_winner"]==self.team_name)&(data_match_08to20["year"]==i)])
                toss_won.append(toss_won_rcb)
                Year.append(i)
                match_won_after_toss=len(data_match_08to20 [(data_match_08to20["toss_winner"]==self.team_name) &( data_match_08to20["winner"]==self.team_name)&(data_match_08to20["year"]==i)])
                toss_match_won_ratio=[match_won_after_toss,toss_won_rcb-match_won_after_toss]
                label_rcb=["Match Won","Match Lost"]
                palette_color = sns.color_palette('Set2')
                plt.pie(toss_match_won_ratio,labels=label_rcb,colors=palette_color, autopct='%.0f%%',explode=(0.1,0))
                plt.title(f"Toss  won by "+self.team_name+" in "+str(i))
                plt.show();
                label=[self.team_name]
                x = np.arange(len(label)) # the label locations
                width = 0.35  # the width of the bars
                fig, ax = plt.subplots()
                rects1 = ax.bar(x - width/2, toss_won_rcb, width, label='Toss won')
                rects2 = ax.bar(x + width/2, match_won_after_toss, width, label='Matches won')
                # Add some text for labels, title and custom x-axis tick labels, etc.
                ax.set_ylabel('Number')
                ax.set_title(f"Toss to matches won ratio of "+self.team_name+" in "+str(i))
                ax.set_xticks(x, label)
                ax.legend()
                ax.bar_label(rects1, padding=3)
                ax.bar_label(rects2, padding=3)
                fig.tight_layout()
                plt.show()
            else:
                print(f"\n Not Playes in Season",i,"\n")

    def compare(self):
        year=self.year
        toss_won=[]
        match_won=[]
        Year=[]  
        
        for i in year:
            

            result =data_match_08to20 [(data_match_08to20["toss_winner"]==self.team_name)&(data_match_08to20["year"]==i)]
            if(i in result.values):
                toss_won_rcb =len(data_match_08to20 [(data_match_08to20["toss_winner"]==self.team_name)&(data_match_08to20["year"]==i)])
                toss_won.append(toss_won_rcb)
                Year.append(i)
                match_won_after_toss=len(data_match_08to20 [(data_match_08to20["toss_winner"]==self.team_name) &( data_match_08to20["winner"]==self.team_name)&(data_match_08to20["year"]==i)])
                match_won.append(match_won_after_toss)
        plt.figure(figsize=(10,6))

        plt.plot(Year, toss_won,linestyle='-.', marker='o',label="Toss Won",markerfacecolor='red', markersize=12, color='orange', linewidth=2,markeredgecolor="yellow")
        plt.plot(Year, match_won,linestyle='-.', marker='o',label="Match Won",markerfacecolor='blue', markersize=12, color='skyblue', linewidth=2,markeredgecolor="violet")


    # Set x and y axes labels
        plt.xlabel('Year')
        plt.ylabel('Toss won')

        plt.title(f'Toss win ratio of {self.team_name} by season')
        plt.legend()
        plt.show()



        
```

```python
class stadium_report:
    def __init__(self,stadium_name,year):
        self.stadium_name=stadium_name
        self.year=year
    def overall_plot(self):
        print(f"Allover Report of ",self.stadium_name)
        eden_garden_win_by_runs=data_match_08to20["result"][(data_match_08to20["venue"]==self.stadium_name)&(data_match_08to20["result"]=="runs")]
        eden_garden_win_by_wickets=data_match_08to20["result"][(data_match_08to20["venue"]==self.stadium_name)&(data_match_08to20["result"]=="wickets")]
        eden_garden_toss_field=data_match_08to20["toss_decision"][(data_match_08to20["venue"]==self.stadium_name)&(data_match_08to20["toss_decision"]=="field")]
        eden_garden_toss_bat=data_match_08to20["toss_decision"][(data_match_08to20["venue"]==self.stadium_name)&(data_match_08to20["toss_decision"]=="bat")]
        eden_match_results=[len(eden_garden_win_by_runs),len(eden_garden_win_by_wickets)]
        eden_toss_results=[len(eden_garden_toss_field),len(eden_garden_toss_bat)]
        label_match=["Batting First","Batting Second"]
        label_toss=["Batting Second","Batting First"]
        palette_color = sns.color_palette('Set2')
        plt.pie(eden_match_results,labels=label_match,colors=palette_color, autopct='%.0f%%',explode=(0.1,0))
        plt.title(f"Winning Team in  "+self.stadium_name+" Overall ")
        plt.show();
        palette_color = sns.color_palette('Set2')
        plt.pie(eden_toss_results,labels=label_toss,colors=palette_color, autopct='%.0f%%',explode=(0.1,0))
        plt.title("Toss Results in "+self.stadium_name)
        plt.show();
    def season_by_season_plot(self):
        year=self.year
        for i in year:
            result=data_match_08to20[(data_match_08to20["venue"]==self.stadium_name)&(data_match_08to20["result"]=="runs")&(data_match_08to20["year"]==i)]
            print(f"For season",i,"Stats are")
            if(i in result.values):
                eden_garden_win_by_runs=data_match_08to20["result"][data_match_08to20["venue"]==self.stadium_name][data_match_08to20["result"]=="runs"][data_match_08to20["year"]==i]
                eden_garden_win_by_wickets=data_match_08to20["result"][data_match_08to20["venue"]==self.stadium_name][data_match_08to20["result"]=="wickets"][data_match_08to20["year"]==i]
                eden_garden_toss_field=data_match_08to20["toss_decision"][data_match_08to20["venue"]==self.stadium_name][data_match_08to20["toss_decision"]=="field"][data_match_08to20["year"]==i]
                eden_garden_toss_bat=data_match_08to20["toss_decision"][data_match_08to20["venue"]==self.stadium_name][data_match_08to20["toss_decision"]=="bat"][data_match_08to20["year"]==i]
                eden_match_results=[len(eden_garden_win_by_runs),len(eden_garden_win_by_wickets)]
                eden_toss_results=[len(eden_garden_toss_field),len(eden_garden_toss_bat)]
                label_match=["Batting First","Batting Second"]
                label_toss=["Batting Second","Batting First"]
                palette_color = sns.color_palette('Set2')
                plt.pie(eden_match_results,labels=label_match,colors=palette_color, autopct='%.0f%%',explode=(0.1,0))
                plt.title("Winning Team in "+self.stadium_name +" in " +str(i))
                plt.show();
                palette_color = sns.color_palette('Set2')
                plt.pie(eden_toss_results,labels=label_toss,colors=palette_color, autopct='%.0f%%',explode=(0.1,0))
                plt.title("Toss Results in "+ self.stadium_name + " in "+ str(i))
                plt.show();
                
            else:
                print(f"The Satdiom ",self.stadium_name," was not useed to play in ",i,"\n")


    
```
